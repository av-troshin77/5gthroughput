## 4. Save Plot

```python
run_number = 10
run = df.loc[(df['run_num']==run_number)][["seq_num", "Throughput"]]
run = run[:120]
run=run.set_index("seq_num")
run.plot(figsize=(10,5),x_compat=True,grid=True)
```

```python
plt.figure(figsize=(10,5))
plt.plot(run, 'k', linewidth=2)
plt.title('Пропускная способность 5G')
plt.ylabel('Мбит/c')
plt.xlabel('сек')
plt.savefig('5g_throughput.png')
plt.show()
```
