# Error Bar Charts of MAE and RMSE

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Plotting MAE

```python
params = ['-', 'L', 'M', 'C']
# mae errors, Mbps
naive_mae = [270.6, 0,     0,     0]
aver_mae = [237.3, 0,     0,     0]
knn_mae =   [215.3, 211.6, 221.9, 248.7]
dtree_mae = [210.4, 210.4, 209.3, 208.9]
rf_mae =    [208.2, 203.3, 205.5, 201.5]
ext_mae =   [220.7, 192.4, 203.6, 197.6]
gb_mae =    [206.4, 191.7, 205.9, 200.0]
mlp_mae =   [191.8, 185.6, 188.6, 188.4]
cnn_mae =   [191.5, 186.6, 187.8, 190.4]
lstm_mae =  [195.4, 185.9, 189.4, 186.9]
```

```python
# plot 
x = np.arange(len(params)) # label locations
width = 0.1  # the width of the bars
fig, ax = plt.subplots(figsize=(10, 5))
rect1 = ax.bar(x-0.3, knn_mae, width, label='kNN', align='center', color='r')
rect2 = ax.bar(x-0.2, dtree_mae, width, label='DTree', align='center', color='b')
rect3 = ax.bar(x-0.1, rf_mae, width, label ='RForest', align='center', color='g')
rect4 = ax.bar(x, ext_mae, width, label ='ExTree', align='center', color='y')
rect5 = ax.bar(x+0.1, gb_mae, width, label ='GBoost', align='center', color='c') 
rect6 = ax.bar(x+0.2, mlp_mae, width, label ='MLP', align='center', color='m')
rect7 = ax.bar(x+0.3, cnn_mae, width, label ='CNN', align='center', color='k')
rect8 = ax.bar(x+0.4, lstm_mae, width, label ='LSTM', align='center', color='0.8')
ax.set_xticks(x)
ax.set_xticklabels(params)
ax.set_title('MAE, Мбит/c')
ax.set_xlabel('Параметры')
ax.legend(loc='upper center', ncol=4)
ax.grid()
ax.set_ylim(0,300)
fig.tight_layout()
# save fig
fig.savefig('MAE_bars_colors1.png')
plt.show()
```

```python
# plot black-white bars
plt.clf() # clear the figure
x = np.arange(len(params)) # label locations
width = 0.1  # the width of the bars
fig, ax = plt.subplots(figsize=(10, 5))
rect1 = ax.bar(x-0.3, knn_mae, width, label='kNN', align='center', color='k',
               edgecolor='k', linewidth=2)
rect2 = ax.bar(x-0.2, dtree_mae, width, label='DTree', align='center', color='w', 
               edgecolor='k', linewidth=2, hatch='-')
rect3 = ax.bar(x-0.1, rf_mae, width, label ='RForest', align='center', color='w', 
               edgecolor='k', linewidth=2, hatch='/')
rect4 = ax.bar(x, ext_mae, width, label ='ExTree', align='center', color='w', 
               edgecolor='k', linewidth=2, hatch='\\')
rect5 = ax.bar(x+0.1, gb_mae, width, label ='GBoost', align='center', color='w',
               edgecolor='k', linewidth=2, hatch='x') 
rect6 = ax.bar(x+0.2, mlp_mae, width, label ='MLP', align='center', color='w', 
               edgecolor='k', linewidth=2, hatch='*')
rect7 = ax.bar(x+0.3, cnn_mae, width, label ='CNN', align='center', color='w',
               edgecolor='k', linewidth=2, hatch='.')
rect8 = ax.bar(x+0.4, lstm_mae, width, label ='LSTM', align='center', color='0.8', 
               edgecolor='k', linewidth=2)
ax.set_xticks(x)
ax.set_xticklabels(params)
ax.set_title('MAE, Мбит/c')
ax.set_xlabel('Параметры')
ax.legend(loc='upper center', ncol=4)
ax.grid()
ax.set_ylim(0,300)
fig.tight_layout()
# save fig
fig.savefig('MAE_bars_back_white1.png')
plt.show()
```
